```python
import numpy as np
import pywt
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Load the image
img = mpimg.imread('02.png')

# Check if the image is RGB and convert to grayscale by averaging channels
if img.ndim == 3:
    img = np.mean(img, axis=2)

# Perform 1-level DWT
coeffs2 = pywt.dwt2(img, 'db2', mode='periodization')
cA, (cH, cV, cD) = coeffs2  # Extracting coefficients

# Perform inverse DWT
imgr = pywt.idwt2(coeffs2, 'db2', mode='periodization')

# Ensure the reconstructed image has the correct data type and range
imgr = np.clip(imgr, 0, 255)
imgr = np.uint8(imgr)

# Plotting the wavelet coefficients and the reconstructed image
plt.figure(figsize=(12, 10))

# Plot approximation coefficients
plt.subplot(2, 3, 1)
plt.imshow(cA, cmap=plt.cm.gray)
plt.title('CA: Approximation Coeff.', fontsize=12)

# Plot horizontal detail coefficients
plt.subplot(2, 3, 2)
plt.imshow(cH, cmap=plt.cm.gray)
plt.title('CH: Horizontal Detailed Coeff.', fontsize=12)

# Plot vertical detail coefficients
plt.subplot(2, 3, 3)
plt.imshow(cV, cmap=plt.cm.gray)
plt.title('CV: Vertical Detailed Coeff.', fontsize=12)

# Plot diagonal detail coefficients
plt.subplot(2, 3, 4)
plt.imshow(cD, cmap=plt.cm.gray)
plt.title('CD: Diagonal Detailed Coeff.', fontsize=12)

# Plot original image
plt.subplot(2, 3, 5)
plt.imshow(img, cmap=plt.cm.gray)
plt.title('Original Image', fontsize=12)



plt.tight_layout()
plt.show()
```
